A stack is a linear data structure that follows the Last-In-First-Out (LIFO) principle.

It means the most recently added element is removed first

e.g. plates stack on wedding ceremony




```python
stack = []

# Push
stack.append('A')
stack.append('B')
stack.append('C')
print("Stack: ", stack)

# Peek
topElement = stack[-1]
print("Peek: ", topElement)

# Pop
poppedElement = stack.pop()
print("Pop: ", poppedElement)

# Stack after Pop
print("Stack after Pop: ", stack)

# isEmpty
isEmpty = not bool(stack)
print("isEmpty: ", isEmpty)

# Size
print("Size: ",len(stack))
```

```python
#make a back button class

class backbutton:
    def __init__(self):
        self.stack=[]
    def visit(self,page):
        self.stack.append(page)
    def back(self):
        if len(self.stack)<1:
            return self.stack[0]
        else:
            self.stack.pop()
            return self.stack[-1]
```
