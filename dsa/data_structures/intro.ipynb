Time complexity → how fast it runs

Space complexity → how much extra memory it needs

```python
# ============================================================
# SPACE AND TIME COMPLEXITY — SINGLE FILE NOTES WITH EXAMPLES
# ============================================================

# ------------------------------------------------------------
# WHAT IS TIME COMPLEXITY?
# ------------------------------------------------------------
# Time complexity describes how the execution time of a program
# grows as the input size (n) increases.
# We express it using Big-O notation.


# ------------------------------------------------------------
# O(1) — CONSTANT TIME
# ------------------------------------------------------------
# Runtime does not depend on input size.

def constant_time(arr):
    return arr[0]  # Always one operation


# ------------------------------------------------------------
# O(n) — LINEAR TIME
# ------------------------------------------------------------
# Runtime grows directly with input size.

def linear_time(arr):
    for x in arr:          # Loop runs n times
        print(x)


# ------------------------------------------------------------
# O(n²) — QUADRATIC TIME
# ------------------------------------------------------------
# Nested loops over the same input.

def quadratic_time(arr):
    for i in arr:          # n
        for j in arr:      # n
            print(i, j)    # n * n operations


# ------------------------------------------------------------
# O(log n) — LOGARITHMIC TIME
# ------------------------------------------------------------
# Input size reduces by a constant factor each iteration.

def logarithmic_time(n):
    while n > 1:
        n //= 2            # Input halves every loop


# ------------------------------------------------------------
# O(n log n) — LINEARITHMIC TIME
# ------------------------------------------------------------
# Common in efficient sorting algorithms.

def linear_log_time(arr):
    arr.sort()             # Python sort is n log n


# ------------------------------------------------------------
# O(2^n) — EXPONENTIAL TIME
# ------------------------------------------------------------
# Each call generates multiple recursive calls.

def exponential_time(n):
    if n <= 1:
        return
    exponential_time(n - 1)
    exponential_time(n - 1)


# ------------------------------------------------------------
# SEQUENTIAL LOOPS — ADD COMPLEXITY
# ------------------------------------------------------------

def sequential_loops(n):
    for i in range(n):     # O(n)
        pass

    for j in range(n):     # O(n)
        pass
    # Total: O(n) + O(n) = O(n)


# ------------------------------------------------------------
# NESTED LOOPS — MULTIPLY COMPLEXITY
# ------------------------------------------------------------

def nested_loops(n):
    for i in range(n):     # O(n)
        for j in range(n): # O(n)
            pass
    # Total: O(n²)


# ------------------------------------------------------------
# MIXED LOOP EXAMPLE
# ------------------------------------------------------------

def mixed_loops(n):
    for i in range(n):     # O(n)
        j = 1
        while j < n:       # O(log n)
            j *= 2
    # Total: O(n log n)


# ------------------------------------------------------------
# WHAT IS SPACE COMPLEXITY?
# ------------------------------------------------------------
# Space complexity measures extra memory used by an algorithm,
# excluding the input itself.


# ------------------------------------------------------------
# O(1) — CONSTANT SPACE
# ------------------------------------------------------------

def constant_space():
    a = 10
    b = 20
    return a + b


# ------------------------------------------------------------
# O(n) — LINEAR SPACE
# ------------------------------------------------------------
# Memory grows with input size.

def linear_space(n):
    arr = []
    for i in range(n):
        arr.append(i)      # Uses extra memory proportional to n
    return arr


# ------------------------------------------------------------
# RECURSION SPACE — CALL STACK
# ------------------------------------------------------------

def recursion_space(n):
    if n == 0:
        return
    recursion_space(n - 1)
    # Stack depth grows to n → O(n) space


# ------------------------------------------------------------
# LOGARITHMIC RECURSION SPACE
# ------------------------------------------------------------

def log_recursion_space(n):
    if n <= 1:
        return
    log_recursion_space(n // 2)
    # Stack depth grows to log n → O(log n)


# ------------------------------------------------------------
# TIME VS SPACE TRADE-OFF
# ------------------------------------------------------------
# Using extra memory to reduce execution time.

def fast_duplicate_check(arr):
    seen = set()           # O(n) extra space
    for x in arr:
        if x in seen:
            return True    # O(1) lookup
        seen.add(x)
    return False
    # Time: O(n), Space: O(n)


def slow_duplicate_check(arr):
    for i in range(len(arr)):
        for j in range(i):
            if arr[i] == arr[j]:
                return True
    return False
    # Time: O(n²), Space: O(1)


# ------------------------------------------------------------
# SUMMARY (MENTAL CHECKLIST)
# ------------------------------------------------------------
# 1. Count loops
# 2. Check if loops are nested
# 3. See if input shrinks (log n)
# 4. Watch recursion depth
# 5. Track extra data structures
# 6. Ignore constants and small terms


# ------------------------------------------------------------
# THIS FILE IS PURELY EDUCATIONAL
# Can be copied into a .py file or used for revision
# ------------------------------------------------------------
```
