```python
arr = [8,3,5,2]
n = len(arr)
for i in range(1, n):
    key = arr[i]
    j = i - 1

    while j >= 0 and arr[j] > key:
        arr[j + 1] = arr[j]
        j -= 1

    arr[j + 1] = key
```

```python
arr = [8,3,5,2]

def binary_search(arr,target):
    n = len(arr)
    left = 0
    right= n-1
    

    while left < right:
        mid = (right+left) // 2
        if target == arr[mid]:
            return mid
        elif target > arr[mid]:
            left=mid+1
        else:
            right = mid -1
```
