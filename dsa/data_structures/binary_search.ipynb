```python
# arr = [1,2,4,5,7]
#target = 5

def binary_search(arr,target):
    left = 0
    right = len(arr)-1

    while left < right:
        mid = (left + right) // 2

        if arr[mid]== target:
            return mid
        elif arr[mid] < target:
            left = mid +1
        else:
            right = mid -1

    return -1
```

```python
arr = [1,2,4,5,7]
target = 5

binary_search(arr,target)
```
