Two Pointers Technique
Last Updated : 13 Feb, 2026
The Two-Pointers Technique is a simple yet powerful strategy where you use two indices (pointers) that traverse a data structure - such as an array, list, or string - either toward each other or in the same direction to solve problems more efficiently

Two pointers is really an easy and effective technique that is typically used for Two Sum in Sorted Arrays, Closest Two Sum, Three Sum, Four Sum, Trapping Rain Water and many other popular interview questions.


```python
1️⃣ What Two Pointers means

Two pointers means using two indices to scan a list efficiently instead of using nested loops.


One rule to remember

If a problem asks about pairs, ranges, or symmetry, think two pointers.
```

```python
# brute force

#Find two numbers that sum to 10.
a=[1,2,3,4,6,8]
target = 10
numbers = []

# def two_sum()len(a)
for i in range(0,6):
    for j in range(1,6):
        if a[i]+a[j] == target:
            numbers.append([a[i],a[j]])

numbers


```

```python
# two pointers

a = [3,1,5,2,7,9,4,10]
target = 10
a = sorted(a)

left,right = 0,len(a)-1

while left < right:
    if a[left] + a[right] == target:
        print(a[left],a[right])
        left +=1
        right+=1
    elif a[left] + a[right] < target:
        left +=1
    else:
        right -=1
        
    
```

```python
# find unique elements

input=[1,1,2,2,3,4,4,5,7,8,3,2,5,2,3,4]

# brute force solution

output = []
for i in input:
    if i in output:
        continue
    output.append(i)

output
```

```python
# two pointers

slow = 0 

input = sorted(input)

for fast in range(1,len(input)):
    if input[fast]!=input[slow]:
        slow +=1
        input[slow]=input[fast]
        

print(input[:slow+1])
```
