```python
#(BMR=(10\times weight\ in\ kg)+(6.25\times height\ in\ cm)-(5\times age\ in\ years)+5\)
```

```python
def maintainance_calories(weight,height,age):
    bmr = 10*weight + 6.25*height - 5*age +5
    m_calories = bmr*1.5
    return m_calories

weight = int(input("enter your weight in kgs:"))
height = int(input("enter your height in centimeters : "))
age = int(input("enter your age in years :"))

maintainance_cal = maintainance_calories(weight,height,age)
print(f"your maintainance_calories are {maintainance_cal} cal")

no_of_chapatis = int(input('no of chapatis did you eat today(including lunch and dinner) :'))
no_of_ricebowls =int(input('no of rice bowl :'))
no_of_dal_bowls = int(input('enter no of bowls of dal :'))
which_subji = int(input('enter 1 for veg 2 for non veg :'))

calories_for_chapati = 90
calories_for_ricebowl = 200
calories_for_dal_bowl = 250
calories_for_veg_subji = 150
calories_for_non_veg_subji = 300

calories_without_subji = no_of_chapatis*calories_for_chapati + no_of_ricebowls*calories_for_ricebowl + no_of_dal_bowls*calories_for_dal_bowl

if which_subji ==1:
    total_daily_intake = calories_without_subji + 2*calories_for_veg_subji
elif which_subji == 2:
    total_daily_intake = calories_without_subji + 2*calories_for_non_veg_subji

print(f'your todays intake is {total_daily_intake} cal')

if total_daily_intake < maintainance_cal:
    print('you are in deficit')
else:
    print('you are in surplus')



```
