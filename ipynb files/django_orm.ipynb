```python
#Django automacity and transaction

 #transactions and atomicity are used to ensure that a group of database operations executes
 #  completely or not at all

# Atomicity means:

# “All operations in a transaction are completed successfully, or none are.”

# If one fails, Django rolls back everything — no partial writes to the database.



def safe_save():
    try:
        with transaction.atomic():
            # first write
            obj1 = Site.objects.create(name='Safe Site', http_port='80', ip_address='10.0.0.1')
            # second write (wrong field causes error)
            obj2 = DeviceModel.objects.create(site=obj1, ip_address='INVALID')  # will fail
    except Exception as e:
        print("Transaction rolled back:", e)


If any statement inside the atomic() block fails (throws an error),
all previous DB operations in that block are rolled back.

with transaction.atomic():
    model1.save()
    model2.save()
    # If this fails, model1 and model2 won't be saved
    model3.save()  
```
