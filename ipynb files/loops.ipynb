```python
num = int(input("Enter a number: "))  # Example: 121
original = num
reverse = 0

while num > 0:
    digit = num % 10
    print(digit)
    reverse = reverse * 10 + digit
    print(reverse)
    num = num // 10
    print(num)

if original == reverse:
    print("Palindrome")
else:
    print("Not a palindrome")
```
