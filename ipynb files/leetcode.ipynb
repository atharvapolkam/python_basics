```python
def romanToInt(s):
    roman = {
            'I': 1,
            'V': 5,
            'X': 10,
            'L': 50,
            'C': 100,
            'D': 500,
            'M': 1000
        }
    
    value=0
    total=0
    prev=0

    for i in reversed(s):
        value=roman[i]
        

        if value < prev:
            total-=value
        else:
            total+=value
            prev=value
    return total



romanToInt('IX')
```

```python
# 238. Product of Array Except Self

# Input: nums = [1,2,3,4]
# Output: [24,12,8,6]

nums = [1,2,3,4]


answer = []

for i in range(len(nums)):
    product=1
    for j in range(len(nums)):
        
        if i != j:
            product=nums[j]*product

    answer.append(product)

print(answer)
```

```python
class Solution:
    def productExceptSelf(self, nums):
        n = len(nums)
        answer = [1] * n

        # Step 1: Left pass (prefix products)
        prefix = 1
        for i in range(n):
            answer[i] = prefix
            prefix *= nums[i]

        # Step 2: Right pass (suffix products)
        suffix = 1
        for i in reversed(range(n)):
            answer[i] *= suffix
            suffix *= nums[i]

        return answer
```
