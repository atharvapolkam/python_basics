```python
Python has a set of built-in methods that you can use on dictionaries.

Method	Description
clear()	Removes all the elements from the dictionary
copy()	Returns a copy of the dictionary
fromkeys()	Returns a dictionary with the specified keys and value
get()	Returns the value of the specified key
items()	Returns a list containing a tuple for each key value pair
keys()	Returns a list containing the dictionary's keys
pop()	Removes the element with the specified key
popitem()	Removes the last inserted key-value pair
setdefault()	Returns the value of the specified key. If the key does not exist: insert the key, with the specified value
update()	Updates the dictionary with the specified key-value pairs
values()	Returns a list of all the values in the dictionary
```

```python
from bisect import bisect_right, bisect_left
flowers=[[1,10],[3,3]]
```

```python
starts = sorted(f[0] for f in flowers)
ends = sorted(f[1] for f in flowers)
```

```python
ends
```

```python
started = bisect_right(starts, t)
ended   = bisect_left(ends, t)
```

```python
people=[2,3,7,11]
```

```python
bloom_count = {}
 
for t in set(people):
    started = bisect_right(starts, t)
    ended   = bisect_left(ends, t)
    bloom_count[t] = started - ended
```

```python
# 1. Frequency Counter

# Input: "abracadabra"
# Output: {'a': 5, 'b': 2, 'r': 2, 'c': 1, 'd': 1}


a= "abracadabra"
b={}

for i in a:
    if i not in b:
        b[i]=1
    else:
        b[i]=b[i]+1
print(b)
from collections import Counter

Counter(a)
```

```python
# Input: {'a': 1, 'b': 2, 'c': 1}
# Output: {1: ['a', 'c'], 2: ['b']}
# 👉 Learn handling duplicate values.
# 2. Invert a Dictionary

a={'a': 1, 'b': 2, 'c': 1}
b={}

for i,j in a.items():
    if j not in b:
        b[j]=[i]
    else:
        b[j].append(i)
b


```

```python
a={'a': 1, 'b': 2, 'c': 1}
from collections import defaultdict
b=defaultdict(list) # default value for any new key is empty list


for i,j in a.items():
    b[j].append(i)
b

# Step by step execution:

# When i='a', j=1

# b[1] is accessed.

# Since key 1 is missing, defaultdict calls list() → creates [].

# Then .append('a') → b[1] = ['a'].

# When i='b', j=2

# b[2] is missing.

# list() is called → [].

# .append('b') → b[2] = ['b'].

# When i='c', j=1

# b[1] already exists (['a']).

# .append('c') → b[1] = ['a', 'c'].
```
