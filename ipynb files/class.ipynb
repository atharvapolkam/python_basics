```python
#  Instance vs Class vs Static Methods


# a) Instance Method (normal)

# Uses self

class Dog:
    def __init__(self, name):
        self.name = name

    def speak(self):  # INSTANCE method
        print(f"{self.name} says woof!")


# Class Method → acts on the class itself

class Dog:
    count = 0

    def __init__(self, name):
        self.name = name
        Dog.count += 1

    @classmethod
    def total_dogs(cls):
        return cls.count

# ✅ c) Static Method → no self, no cls

# Utility methods that relate to the class but don’t touch its data

class Math:
    @staticmethod
    def add(x, y):
        return x + y
```

```python
# ✅ 3. Encapsulation

# Hiding internal variables from direct access

# In Python: use _ (protected) or __ (private-like)

class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # "private"

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance

# 🔒 __balance can’t be accessed directly: acc.__balance will give error
# ✅ Must use get_balance() method


acc = BankAccount(500)

```

```python
acc.get_balance()
```

```python
# ✅ 4. Inheritance (child class gets parent’s features)
class Animal:
    def speak(self):
        print("Animal sound")

class Dog(Animal):  # Dog inherits from Animal
    def speak(self):
        print("Woof")

d = Dog()
d.speak()  # Output: Woof
```

```python
# ✅ 5. polymorphism
animals = [Dog(), Animal()]

for a in animals:
    a.speak()  # Calls Dog.speak() or Animal.speak() depending on type
```

```python
'''Problem 1:
Create a class called Student that has:

Attributes: name, roll_no

Method: display() — prints Name: <name>, Roll No: <roll_no>

Task:

Create the class.

Create two objects of the class (student1, student2).

Call the display() method for both.'''


class Student():
    def __init__(self,name,roll_no):
        self.name= name
        self.roll_no = roll_no

    def display(self):
        return f"name is {self.name} and roll no is {self.roll_no}"

student1 = Student('a',5)
student2 = Student('b', 6)

print(student2.display())
```

```python
'''Problem 2:
Create a class called Rectangle that has:

Attributes: length, breadth

Methods:

area() → returns the area of the rectangle.

perimeter() → returns the perimeter of the rectangle.

Task:

Create the class.

Create one rectangle object with length=4, breadth=3.

Print the area and perimeter.'''


class Rectangle :
    def __init__(self,length,breadth):
        self.length = length
        self.breadth = breadth
    def area(self):
        return self.length*self.breadth
    def perimeter(self):
        return 2*self.length + 2*self.breadth
    
r1 = Rectangle(4,5)
print(r1.area())
print(r1.perimeter())
```

```python
############### INHERITENCE ###################



'''Create two classes:

Parent Class: Employee

Attributes: name, salary

Method: show_details() → prints name and salary

Child Class: Manager

Additional Attribute: department

Method: show_manager() → prints name, salary, and department

Task:
Create an object of Manager and call both methods.'''



class Employee :
    def __init__(self,name ,salary):
        self.name = name
        self.salary = salary

    def show_details(self):
        return self.name , self.salary
    
class Manager(Employee):
    def __init__(self, name, salary, department):
        self.name = name 
        self.salary= salary
        self.department = department
    
    def show_manager(self):
        return self.name,self.salary,self.department
    
c1 = Manager('a',5,'hr')


print(c1.name)              # Output: a
print(c1.salary)            # Output: 5
print(c1.department)        # Output: HR
print(c1.show_manager())    # Output: ('a', 5, 'HR')
print(c1.show_details())

        
```

```python
'''✅ Problem: Simple Inventory Manager
Create a class Inventory with the following methods:

add_item(item_name, quantity)

remove_item(item_name, quantity)

list_items() → returns a dictionary of all items and their quantities.'''


class Inventory:
    def __init__(self):
        self.items = {}

    
    def add_item(self,item,quantity):
        if item in self.items:
            self.items[item]+=quantity
        else:
            self.items[item]=quantity
        return self.items
    
    def remove_item(self,item,quantity):
        if item not in self.items:
            print("item is not available in your inventory")
        elif item in self.items and quantity <= self.items[item]:
            self.items[item]-= quantity
            if self.items[item] == 0:
                self.items.pop(item)

    def list_items(self):
        return self.items
    
```
