```python
#Q7 Write a Python function that merges all overlapping intervals and 
# returns a list of non-overlapping intervals covering all ranges.

# Input

intervals = [[1,3],[2,6],[8,10],[15,18]]

output=[]

# Output

# [[1,6],[8,10],[15,18]]

for i in range(0,3):
    for j in range(1,3):
        if intervals[i][1]<intervals[j][0]:
            output.append([intervals[i][0],intervals[j][1]])
        else:
            output.append(intervals[i])
        if (j-i) !=1:
            break

output
```

```python
#q3  Write a program that groups words based on their length.

from collections import defaultdict


words = ["cat", "dog", "elephant", "bat", "tiger", "lion"]

# output
# {
# 3: ['cat', 'dog', 'bat'],
# 5: ['tiger'],
# 4: ['lion'],
# 8: ['elephant']
# }

output = defaultdict(list)

for i in words:
    output[len(i)].append(i)

output
    
```

```python
# Q 6

# Problem: Shape Area Calculator (OOP)

# Create a program using Object-Oriented Programming that demonstrates:

# Encapsulation,Inheritance,Polymorphism

# Requirements

# Create a base class Shape.

# It should contain a private attribute name.

# Provide getter method to access the name.

# Define a method:

# area()

# This method will be overridden in subclasses.

# Create two subclasses that inherit from Shape.

# Rectangle

# Attributes:

# length, width

# Area formula:
# length * width


# Circle

# Attribute:
# radius

# Area formula:

# 3.14 * radius * radius

# Demonstrate polymorphism

# Write a loop that takes different shape objects and prints their areas using the same method call.

# Example Usage
# Rectangle area: 20
# Circle area: 78.5
```

```python
import math


class shape :
    __name = "hello "

    # polymorphism
    def area(self,length,width):
        print((length* width))
    
    def area(self,radius):
        print(math.pi * math.pow(radius,2))

class Rectange(shape):

    def area(self,length,width):
        print((length* width))

class Circle(shape):
    def area(self,radius):
        print(math.pi * math.pow(radius,2))

lst = ["rect","circle","rect","circle",]
length , width , radius = map (int, input("Emter the length, width, radius seperated by space".split()))
for i in lst :
    if i  == 'rect':
        c =  Circle()
        c.area(radius)
    elif( i == 'circle'):
        R = Rectange()
        R.area(length,width)
    else :
        print("Invalid")        
```
