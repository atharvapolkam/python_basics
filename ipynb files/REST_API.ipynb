```python
# What is a REST API?

# A REST API (Representational State Transfer) is a set of rules for 
# communication between a client (like your web app or mobile app) and a server.
```

```python
| Method | Purpose             | Example                  |
| ------ | ------------------- | ------------------------ |
| GET    | Retrieve data       | Get list of users        |
| POST   | Create new data     | Add a new user           |
| PUT    | Update all fields   | Update user info         |
| PATCH  | Update part of data | Change users email only  |
| DELETE | Remove data         | Delete a user            |
```
