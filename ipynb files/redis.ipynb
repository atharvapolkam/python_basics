REDIS

REMOTE DICTIONARY SERVER

It’s an in-memory database → stores data in RAM (super fast).

Often used for:

Caching (store temporary results so you don’t re-compute).

Queues (task processing).

Pub/Sub (messaging).

Session storage (user login data, tokens, etc.).

```python
import redis

# Connect to Redis
r = redis.Redis(host='localhost', port=6379, db=0)

# Strings
r.set("name", "atharva")
print(r.get("name"))  # b'atharva'

# Lists
r.lpush("tasks", "task1")
r.lpush("tasks", "task2")
print(r.lrange("tasks", 0, -1))  # [b'task2', b'task1']

# Hash
r.hset("user:1", mapping={"name": "atharva", "age": "25"})
print(r.hgetall("user:1"))  # {b'name': b'atharva', b'age': b'25'}
```
