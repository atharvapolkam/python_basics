```python


def removeDuplicates(nums):
    """
    :type nums: List[int]
    :rtype: int
    """
    l=[]

    for i in nums:
        if i in l:
            continue
        else:
            l.append(i)
    return len(l) , l

removeDuplicates([1,1,1,3,4,4,5,5,6,7,8,9,9])
```

```python
def find(arr, target):
    # Validate input types
    if not isinstance(target,int):
        raise TypeError
    if not isinstance(arr,list) or not all(isinstance(i,int) for i in arr):
        raise TypeError
    
    for i in arr:
        if i==target:
            return True
        
    return False

find([1,2,4,5],7)
```

```python
# Input: candies = [2,3,5,1,3], extraCandies = 3
# Output: [true,true,true,false,true] 
# Explanation: If you give all extraCandies to:
# - Kid 1, they will have 2 + 3 = 5 candies, which is the greatest among the kids.
# - Kid 2, they will have 3 + 3 = 6 candies, which is the greatest among the kids.
# - Kid 3, they will have 5 + 3 = 8 candies, which is the greatest among the kids.
# - Kid 4, they will have 1 + 3 = 4 candies, which is not the greatest among the kids.
# - Kid 5, they will have 3 + 3 = 6 candies, which is the greatest among the kids.
```

```python
candies = [2,3,5,1,3]
extraCandies = 3
p=[]

for i in candies:
    if int(i+extraCandies) >= max(candies):
        p.append(True)
    else:
        p.append(False)
print(p)

```

```python
# Input: s = "IceCreAm"

# Output: "AceCreIm"

# Explanation:

# The vowels in s are ['I', 'e', 'e', 'A']. On reversing the vowels, s becomes "AceCreIm".



s = "IceCreAm"

#  I c e C r e A m
#   A   C r     m
#
#[A , e ,e ,I]

# main code starts from below


vowels= ['a','e','i','o','u']

new_list = []

vowel_list=[]
index_list = []

d={}

for i,j in enumerate(s) :
    if j.lower() not  in vowels:
        new_list.append(j)
    elif j.lower() in vowels:
        vowel_list.append(j)
        index_list.append(i)


vowel_list.reverse()

for k in range(len(index_list)):
    id = index_list[k]
    s = s[:id] + vowel_list[k] + s[id+1:]

return s

```
