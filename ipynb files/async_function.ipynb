```python
# 1. What does await asyncio.sleep(seconds) do?

# It pauses the current async function for a given number of seconds without blocking the entire program.

# await → Tells Python: “Pause here and come back when this is done.”

# asyncio.sleep(x) → Waits for x seconds asynchronously (without blocking the CPU).

```

```python
# Concept	Explanation
# async def	Defines an asynchronous function
# await	Pauses until the awaited task is done, allows others to run meanwhile
# asyncio.run()	Starts the event loop to run the top-level coroutine
# asyncio.sleep	Like time.sleep, but doesn’t block
# asyncio.gather	Runs multiple async tasks together
```

```python
import asyncio
async def my_task(name,time):
    print(f'{name} running ')
    await asyncio.sleep(time)
    print(f'task {name} is completed in {time}')
    await asyncio.sleep(1)
    print(f"abc,{name}")

async def main():
    asyncio.gather(
        my_task('dinner',3),
        my_task('call',1),
        my_task('grind',6),
        my_task('grind1',6),
        my_task('grind2',6),
        my_task('grind3',7)


    )
await main()
```

```python
import asyncio
import time

def slow_add(x, y):
    time.sleep(2)
    return x + y

async def main():
    loop = asyncio.get_running_loop()
    task = loop.run_in_executor(None, slow_add, 3, 4)

    print("Submitted task:", task)   
    print("Doing other work...")
    await asyncio.sleep(1)
    print("Still working...")

    result = await task
    print("Result:", result)

await main()
```

```python
# Write a coroutine download(name, time) that prints start/end messages for simulated downloads using await asyncio.sleep().

import asyncio

async def download(name, time):
    print(f"Download is starting for {name}")
    await asyncio.sleep(time)
    print(f"Download finished for {name}")

async def main():
    await asyncio.gather(
        download('a', 4),
        download('b', 2),
        download('c', 3)
    )

await main()
```

```python
# Write two coroutines — one prints numbers every second, another prints letters every 2 seconds.
# Run both with asyncio.gather().

import asyncio

async def print_numbers():
    for i in range(1, 6):
        print(i)
        await asyncio.sleep(1)

async def print_letters():
    for c in ['A', 'B', 'C', 'D', 'E']:
        print(c)
        await asyncio.sleep(2)

async def main():
    await asyncio.gather(print_numbers(), print_letters())

await main()
```
