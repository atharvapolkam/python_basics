```python
🔹 Analogy of ThreadPool

# Think of it like a kitchen:

# ThreadPoolExecutor(max_workers=16) = 16 chefs.

# loop.run_in_executor(..., cook_dish, order) = place an order with ingredients.

# One chef takes it, cooks it.

# The waiter (asyncio loop) doesn’t stand idle — he serves other tables.

# When the chef finishes, the waiter picks up the dish (detections) and delivers it.
```

```python
Semaphore

# A semaphore is like a counter + lock that controls how many tasks can run at the same time.


🔹 Why use it in asyncio?

# Without a semaphore:

# If you have 10,000 images, your code might try to run 10,000 async requests at once.

# That will crash memory / network / GPU.

# With Semaphore(500):

# At most 500 tasks run in parallel.

# Others wait their turn.
```
