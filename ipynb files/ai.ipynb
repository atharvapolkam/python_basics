```python
TensorRT is an optimized inference library and toolkit developed by NVIDIA
to maximize the performance (speed and efficiency) of deep learning models on NVIDIA GPUs.

How TensorRT Works
1. Model Import/Export: Trained models are imported from PyTorch, TensorFlow, ONNX, or Caffe. TensorRT parses these models for further processing.

2. Optimization Phase: During this phase, TensorRT

Combines compatible layers to reduce memory copy and computation overhead
Selects the fastest kernels for each layer and hardware
Converts layers and weights to lower-precision formats if specified
Removes redundant layers and recalibrates data flows for efficiency
3. Engine Generation & Deployment: An optimized, lightweight runtime “engine” is generated. This engine is then deployed and used for high-performance inference on NVIDIA GPUs.

Implementation
TensorRT implementation involves converting a trained model into an optimized engine using model parsers.
It applies optimizations like layer fusion, precision calibration (FP16/INT8) and kernel tuning. The engine is then serialized 
and used for fast, low-latency inference on NVIDIA GPUs or edge devices. APIs in Python and C++ make integration easy.
```

```python
threading vs multiprocessing

Multiple threads inside a single process
Share same memory
Lightweight (fast to create)
Good for switching between tasks
    

👉 Multiprocessing = multiple independent processes

Each process has its own memory
Runs on different CPU cores
True parallel execution


#threading example
import threading
import time

def fetch_camera(camera_id):
    print(f"Fetching from camera {camera_id}")
    time.sleep(2)  # simulate network delay
    print(f"Done camera {camera_id}")

threads = []

for i in range(5):
    t = threading.Thread(target=fetch_camera, args=(i,))
    t.start()
    threads.append(t)

for t in threads:
    t.join()



#multiprocess example

from multiprocessing import Process
import time

def heavy_task(n):
    print(f"Processing {n}")
    total = 0
    for i in range(10_000_000):
        total += i
    print(f"Done {n}")

processes = []

for i in range(4):
    p = Process(target=heavy_task, args=(i,))
    p.start()
    processes.append(p)

for p in processes:
    p.join()
```
