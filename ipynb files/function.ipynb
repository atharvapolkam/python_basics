```python
def is_even(n: int) -> bool:
    return n % 2 == 0

The function returns either True or False.

The -> bool tells you the return value should be of type bool.
```

```python
def prime(x):# these is parameter
    for i in range(2,x):
        if x%i==0:
            return False
        elif x==2:
            return True
        else:
            return True
prime(5)# these is argument
```
