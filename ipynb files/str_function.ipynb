````python
🔁 *Top 10 Python String Methods You Should Know 🧵🐍*
 
```
s = "hello world"
```
 
*1. `upper()`* – *Convert to uppercase*  
› `s.upper()` → `"HELLO WORLD"`
 
*2. `lower()`* – *Convert to lowercase*  
› `s.lower()` → `"hello world"`
 
*3. `capitalize()`* – *Capitalize first letter*  
› `s.capitalize()` → `"Hello world"`
 
*4. `title()`* – *Capitalize first letter of each word*  
› `s.title()` → `"Hello World"`
 
*5. `strip()`* – *Remove leading/trailing spaces*  
› `"  hello  ".strip()` → `"hello"`
 
*6. `replace()`* – *Replace substring*  
› `s.replace("world", "Python")` → `"hello Python"`
 
*7. `split()`* – *Split into list*  
› `s.split()` → `["hello", "world"]`
 
*8. `join()`* – *Join list into string*  
› `" ".join(["Python", "is", "fun"])` → `"Python is fun"`
 
*9. `find()`* – *Find first occurrence*  
› `s.find("o")` → `4`
 
*10. `startswith()` / `endswith()`* – *Check prefix/suffix*  
› `s.startswith("hello")` → `True`  
› `s.endswith("world")` → `True`
 
💡 *Bonus Tip:* Use `in` to check if a substring exists → `"Python" in s`
 
````

````python
#a[start : end : step]

a="asdfas  fas  "

strip=a.strip() #The strip() method removes any whitespace from the beginning or the end:
print(strip)

replace=a.replace('s','p')
print(replace)

split=a.split('s')
print(split)

join=" ".join(a)
print(join)
````
